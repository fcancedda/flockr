```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

def tsne(data, n=1, perp=15, lr=5):
    return TSNE(
        n_components=n,
        learning_rate=lr,
        perplexity=perp,
        init='random').fit_transform(data)
def pca(data, n):
    return PCA(n_components=n).fit_transform(data)
```

```python
x1 = np.random.normal(10, scale =.1,size=50)
y1 = np.random.normal(10, scale =.1,size=50)

x2 = np.random.normal(20, scale =.1, size=50)
y2 = np.random.normal(10, scale =.1, size=50)

x3 = np.random.normal(25, scale =1,size=50)
y3 = np.random.normal(10, scale =1,size=50)

x = np.concatenate((x1, x2, x3))
y = np.concatenate((y1,y2, y3))
sns.scatterplot(x, y)
# plt.xlim(5,30)
# plt.ylim(0, 25)

```

```python
X = np.stack((x,y), axis=1)
X
```

```python
tsne_emb = tsne(X, n=2, perp=5, lr=10)
sns.scatterplot(tsne_emb[:,0], tsne_emb[:,1])
```

```python
pca_emb = pca(X, n=2)

sns.scatterplot(pca_emb[:,0], pca_emb[:,1])
```

```python
from mpl_toolkits.mplot3d import Axes3D
import ipympl
%matplotlib widget

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
points_per_cluster = 100

samples = [
    [sample(20, 18, 0, 0.0) for _ in range(points_per_cluster)],
    [sample(20, 12, 0.0, 0.0) for _ in range(points_per_cluster)],
    [sample(20, 18, 0, 0) for _ in range(points_per_cluster)]
]
x_vals = []
y_vals = []
z_vals = []
for i in range(len(samples)):
    xs, ys = zip(*samples[i])
    xs = tuple(x+i%2*40 for x in xs)
    ys = tuple(y+i*10 for y in ys)
    zs = np.random.normal(i + i%2*27, scale =1,size=points_per_cluster)
    # plt.scatter(xs, ys)
    if i==2:
        ax.scatter3D(zs, xs, ys)
        x_vals.append(zs)
        y_vals.append(xs)
        z_vals.append(ys)
    else:
        sns.scatterplot(xs, ys, zs)
        x_vals.append(xs)
        y_vals.append(ys)
        z_vals.append(zs)




# plt.axis("equal")
# plt.xlim(-10, 10)
# plt.ylim(-15, 10)
plt.show()

x = np.concatenate(x_vals)
y = np.concatenate(y_vals)
z = np.concatenate(z_vals)
```

```python
plt.clf()

emb = pca(np.stack((x, y, z), axis=1), n=2)
for i in range(0, len(emb), points_per_cluster):
    cluster = emb[i: i + points_per_cluster]
    plt.scatter(cluster[:,0], cluster[:,1])
plt.axis('equal')
plt.show()
```

```python
plt.clf()

emb = tsne(np.stack((x, y, z), axis=1), n=2, perp=10, lr=5)
for i in range(0, len(emb), points_per_cluster):
    cluster = emb[i: i + points_per_cluster]
    plt.scatter(cluster[:,0], cluster[:,1])
plt.axis('equal')
plt.show()
```
